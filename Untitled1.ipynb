```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from matplotlib.cm import get_cmap
from tqdm import tqdm

# Set random seed for reproducibility
np.random.seed(42)
torch.manual_seed(42)

# Common parameters for both methods
nx = 200       # Number of grid points
L = 1.0        # Domain length [0,L]
gamma = 1.4    # Ratio of specific heats for ideal gas
tmax = 0.2     # Maximum simulation time

# Common parameter for visualizations
time_points = [0.0, 0.04, 0.08, 0.12, 0.16, 0.2]  # Time points for comparison

#############################################
# PART 1: NUMERICAL SOLUTION (FINITE VOLUME)
#############################################

def numerical_solution():
    print("Running numerical solution with Lax-Friedrichs scheme...")

    # Numerical method parameters
    cfl = 0.5      # CFL number for stability
    nt_max = 5000  # Maximum number of time steps

    # Initialize grid for numerical method
    dx = L / nx
    x_num = np.linspace(0, L, nx+1)[:-1]  # Grid points [0,L) for periodic BC

    # Sod shock tube initial conditions
    rho = np.ones_like(x_num)
    u = np.zeros_like(x_num)
    p = np.ones_like(x_num)

    # Set right state
    right_idx = x_num >= 0.5
    rho[right_idx] = 0.125
    p[right_idx] = 0.1

    # Compute initial conserved variables
    E = p / ((gamma - 1) * rho) + 0.5 * u**2  # Specific total energy
    Q = np.zeros((3, nx))  # Conserved variables [rho, rho*u, rho*E]
    Q[0, :] = rho
    Q[1, :] = rho * u
    Q[2, :] = rho * E

    # Store results at different time steps
    results = {0.0: {"rho": rho.copy(), "u": u.copy(), "p": p.copy()}}
    next_save_idx = 1

    # Function to compute fluxes
    def compute_flux(Q):
        rho = Q[0, :]
        u = Q[1, :] / (rho + 1e-10)  # Add small value to avoid division by zero
        E = Q[2, :] / (rho + 1e-10)
        p = (gamma - 1) * rho * (E - 0.5 * u**2)
        p = np.maximum(p, 1e-10)  # Ensure positive pressure

        F = np.zeros_like(Q)
        F[0, :] = rho * u
        F[1, :] = rho * u**2 + p
        F[2, :] = u * (rho * E + p)

        return F

    # Function to compute wavespeeds for artificial viscosity
    def compute_wavespeeds(Q):
        rho = Q[0, :]
        u = Q[1, :] / (rho + 1e-10)
        E = Q[2, :] / (rho + 1e-10)
        p = (gamma - 1) * rho * (E - 0.5 * u**2)
        p = np.maximum(p, 1e-10)

        c = np.sqrt(gamma * p / (rho + 1e-10))  # Sound speed
        return np.abs(u) + c

    # Implement Lax-Friedrichs scheme with artificial viscosity for shock capturing
    def lax_friedrichs_step(Q, dt):
        F = compute_flux(Q)
        wavespeeds = compute_wavespeeds(Q)
        max_wavespeed = np.max(wavespeeds)

        Q_new = np.zeros_like(Q)

        # Interior points
        for i in range(1, nx-1):
            # Lax-Friedrichs flux with artificial viscosity
            flux_diff = (F[:, i+1] - F[:, i-1]) / (2*dx)
            artificial_viscosity = max_wavespeed * (Q[:, i+1] - 2*Q[:, i] + Q[:, i-1]) / (2*dx)
            Q_new[:, i] = Q[:, i] - dt * flux_diff + dt * artificial_viscosity

        # Boundary treatment (transmissive)
        Q_new[:, 0] = Q[:, 0] - dt * (F[:, 1] - F[:, 0]) / dx
        Q_new[:, nx-1] = Q[:, nx-1] - dt * (F[:, nx-1] - F[:, nx-2]) / dx

        return Q_new

    # Time stepping
    t = 0
    nt = 0

    # Progress bar for numerical solution
    pbar = tqdm(total=len(time_points)-1, desc="Numerical solution progress")

    while t < tmax and nt < nt_max:
        # Compute time step based on CFL condition
        rho = Q[0, :]
        u = Q[1, :] / (rho + 1e-10)
        E = Q[2, :] / (rho + 1e-10)
        p = (gamma - 1) * rho * (E - 0.5 * u**2)
        p = np.maximum(p, 1e-10)  # Ensure positive pressure
        c = np.sqrt(gamma * p / (rho + 1e-10))  # Speed of sound
        dt = cfl * dx / np.max(np.abs(u) + c)

        # Make sure we don't overshoot the next time point
        if next_save_idx < len(time_points):
            dt = min(dt, time_points[next_save_idx] - t + 1e-10)

        # Update solution using Lax-Friedrichs
        Q = lax_friedrichs_step(Q, dt)

        # Update time
        t += dt
        nt += 1

        # Extract primitive variables
        rho = Q[0, :]
        u = Q[1, :] / (rho + 1e-10)
        E = Q[2, :] / (rho + 1e-10)
        p = (gamma - 1) * rho * (E - 0.5 * u**2)
        p = np.maximum(p, 1e-10)  # Ensure positive pressure

        # Save results at specified time points
        if next_save_idx < len(time_points) and t >= time_points[next_save_idx] - 1e-10:
            results[time_points[next_save_idx]] = {"rho": rho.copy(), "u": u.copy(), "p": p.copy()}
            next_save_idx += 1
            pbar.update(1)

        # Check for NaN values (indicates instability)
        if np.isnan(Q).any():
            print("NaN detected in solution. Simulation unstable.")
            break

        # Stop if we've saved all requested time points
        if next_save_idx >= len(time_points):
            break

    pbar.close()
    print(f"Numerical solution completed in {nt} time steps")

    return x_num, results

# Get the exact initial condition (for comparison)
def get_exact_initial_condition():
    x = np.linspace(0, L, nx)

    # Initialize with left state
    rho = np.ones_like(x)
    u = np.zeros_like(x)
    p = np.ones_like(x)

    # Set right state
    right_idx = x >= 0.5
    rho[right_idx] = 0.125
    p[right_idx] = 0.1

    return x, {"rho": rho, "u": u, "p": p}

##############################################
# PART 2: PINN SOLUTION (PHYSICS-INFORMED NN)
##############################################

class EulerPINN(nn.Module):
    def __init__(self, hidden_layers=4, neurons=50):
        super(EulerPINN, self).__init__()

        # Neural network for approximating solution
        layers = []
        layers.append(nn.Linear(2, neurons))  # Input: (x, t)
        layers.append(nn.Tanh())

        for _ in range(hidden_layers):
            layers.append(nn.Linear(neurons, neurons))
            layers.append(nn.Tanh())

        layers.append(nn.Linear(neurons, 3))  # Output: (rho, u, p)

        self.net = nn.Sequential(*layers)

        # Initialize weights to get better results
        for m in self.net.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x, t):
        # Combine inputs into single tensor
        inputs = torch.cat([x, t], dim=1)
        outputs = self.net(inputs)

        # Apply constraints to ensure physically meaningful outputs
        rho = torch.exp(outputs[:, 0:1])  # Ensure density is positive
        u = outputs[:, 1:2]  # Velocity can be positive or negative
        p = torch.exp(outputs[:, 2:3])  # Ensure pressure is positive

        return rho, u, p

    def pde_residual(self, x, t):
        # Make inputs require gradients
        x.requires_grad_(True)
        t.requires_grad_(True)

        # Get solution from neural network
        rho, u, p = self.forward(x, t)

        # Energy
        E = p / ((gamma - 1) * rho) + 0.5 * u**2

        # Calculate derivatives using autograd
        # Derivative wrt time
        drho_dt = torch.autograd.grad(rho, t, grad_outputs=torch.ones_like(rho), create_graph=True)[0]
        du_dt = torch.autograd.grad(u, t, grad_outputs=torch.ones_like(u), create_graph=True)[0]
        dp_dt = torch.autograd.grad(p, t, grad_outputs=torch.ones_like(p), create_graph=True)[0]
        dE_dt = torch.autograd.grad(E, t, grad_outputs=torch.ones_like(E), create_graph=True)[0]

        # Derivative wrt space
        drho_dx = torch.autograd.grad(rho, x, grad_outputs=torch.ones_like(rho), create_graph=True)[0]
        du_dx = torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]
        dp_dx = torch.autograd.grad(p, x, grad_outputs=torch.ones_like(p), create_graph=True)[0]
        dE_dx = torch.autograd.grad(E, x, grad_outputs=torch.ones_like(E), create_graph=True)[0]

        # Conservation of mass
        f1 = drho_dt + drho_dx * u + rho * du_dx

        # Conservation of momentum
        f2 = rho * du_dt + rho * u * du_dx + dp_dx

        # Conservation of energy
        f3 = rho * dE_dt + rho * u * dE_dx + p * du_dx

        return f1, f2, f3

    def initial_condition_residual(self, x):
        t = torch.zeros_like(x)
        rho, u, p = self.forward(x, t)

        # Sod shock tube initial conditions
        rho_init = torch.ones_like(rho)
        u_init = torch.zeros_like(u)
        p_init = torch.ones_like(p)

        mask = (x >= 0.5).float()
        rho_init = rho_init * (1 - mask) + 0.125 * mask
        p_init = p_init * (1 - mask) + 0.1 * mask

        return (rho - rho_init), (u - u_init), (p - p_init)


def pinn_solution():
    print("Running PINN solution...")

    # Create model
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = EulerPINN(hidden_layers=4, neurons=64).to(device)

    # Training parameters
    epochs = 10000
    lr = 0.001

    # Optimizer
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=500, factor=0.5, verbose=True)

    # Create training points
    n_points_x = 100
    n_points_t = 50
    n_points_ic = 200

    # Domain points (x, t) for PDE residuals
    x_domain = torch.linspace(0, L, n_points_x, device=device).view(-1, 1)
    t_domain = torch.linspace(0, tmax, n_points_t, device=device).view(-1, 1)
    X, T = torch.meshgrid(x_domain.squeeze(), t_domain.squeeze(), indexing='ij')
    x_collocation = X.reshape(-1, 1)
    t_collocation = T.reshape(-1, 1)

    # Initial condition points (x, t=0)
    x_ic = torch.linspace(0, L, n_points_ic, device=device).view(-1, 1)

    # Training loop
    loss_history = []
    ic_loss_history = []
    pde_loss_history = []
    best_loss = float('inf')
    best_model_state = None

    # Progress bar for PINN training
    pbar = tqdm(range(epochs), desc="PINN training progress")

    for epoch in pbar:
        model.train()
        optimizer.zero_grad()

        # Sample random collocation points for each epoch
        idx = torch.randperm(len(x_collocation))[:200]
        x_batch = x_collocation[idx]
        t_batch = t_collocation[idx]

        # PDE residuals
        f1, f2, f3 = model.pde_residual(x_batch, t_batch)
        pde_loss = torch.mean(f1**2 + f2**2 + f3**2)

        # Initial condition residuals
        ic_rho, ic_u, ic_p = model.initial_condition_residual(x_ic)
        ic_loss = torch.mean(ic_rho**2 + ic_u**2 + ic_p**2)

        # Total loss - give much higher weight to initial condition loss
        # This helps ensure the initial condition is matched more closely
        loss = pde_loss + 50.0 * ic_loss  # Increased weight from 10.0 to 50.0

        # Backward and optimize
        loss.backward()
        optimizer.step()

        # Save loss history
        loss_val = loss.item()
        loss_history.append(loss_val)
        ic_loss_history.append(ic_loss.item())
        pde_loss_history.append(pde_loss.item())

        # Save best model
        if loss_val < best_loss:
            best_loss = loss_val
            best_model_state = model.state_dict().copy()

        # Update learning rate based on loss
        scheduler.step(loss_val)

        # Update progress bar
        if epoch % 100 == 0:
            pbar.set_postfix({"loss": f"{loss_val:.5f}", "ic_loss": f"{ic_loss.item():.5f}", "pde_loss": f"{pde_loss.item():.5f}"})

    pbar.close()
    print(f"PINN training completed with best loss: {best_loss:.5f}")

    # Load best model
    model.load_state_dict(best_model_state)
    model.eval()

    # Generate solution at time points for comparison
    x_eval = torch.linspace(0, L, nx, device=device).view(-1, 1)
    results = {}

    for t_val in time_points:
        t_eval = torch.ones_like(x_eval, device=device) * t_val

        with torch.no_grad():
            rho, u, p = model(x_eval, t_eval)

        results[t_val] = {
            "rho": rho.cpu().numpy().squeeze(),
            "u": u.cpu().numpy().squeeze(),
            "p": p.cpu().numpy().squeeze()
        }

    return x_eval.cpu().numpy().squeeze(), results, loss_history, ic_loss_history, pde_loss_history


#############################################
# PART 3: COMBINED PLOTTING AND ANALYSIS
#############################################

# Get the exact initial condition
x_exact, exact_ic = get_exact_initial_condition()

# Run both solutions
x_pinn, pinn_results, pinn_loss_history, ic_loss_history, pde_loss_history = pinn_solution()
x_num, num_results = numerical_solution()

# Plot loss histories for PINN
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.semilogy(pinn_loss_history)
plt.xlabel('Epoch')
plt.ylabel('Total Loss')
plt.title('PINN Total Training Loss')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.semilogy(ic_loss_history)
plt.xlabel('Epoch')
plt.ylabel('Initial Condition Loss')
plt.title('PINN Initial Condition Loss')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.semilogy(pde_loss_history)
plt.xlabel('Epoch')
plt.ylabel('PDE Loss')
plt.title('PINN PDE Residual Loss')
plt.grid(True)

plt.tight_layout()
plt.savefig("pinn_loss_history_detailed.png", dpi=300, bbox_inches="tight")
plt.close()

# Detailed analysis of initial condition match
variables = ["rho", "u", "p"]
titles = ["Density (Ï)", "Velocity (u)", "Pressure (p)"]

# Plot initial condition comparison specifically (all three variables)
plt.figure(figsize=(15, 5))

for var_idx, (var, title) in enumerate(zip(variables, titles)):
    plt.subplot(1, 3, var_idx + 1)

    # Plot exact initial condition
    plt.plot(x_exact, exact_ic[var], 'k--', linewidth=2, label='Exact IC')

    # Plot numerical solution at t=0
    plt.plot(x_num, num_results[0.0][var], 'b-', linewidth=2, label='FVM (t=0)')

    # Plot PINN solution at t=0
    plt.plot(x_pinn, pinn_results[0.0][var], 'r-', linewidth=2, label='PINN (t=0)')

    plt.xlabel("x", fontsize=12)
    plt.ylabel(title, fontsize=12)
    plt.title(f"Initial Condition: {title}", fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)

plt.tight_layout()
plt.savefig("initial_condition_comparison.png", dpi=300, bbox_inches="tight")
plt.close()

# Create side-by-side plots for each variable (rho, u, p) at each time step
colors = ['blue', 'red', 'green']

# 1. Individual comparison plots for each variable at each time step
for var_idx, (var, title, color) in enumerate(zip(variables, titles, colors)):
    for t_idx, t_val in enumerate(time_points):
        plt.figure(figsize=(10, 6))

        # Plot both solutions
        plt.plot(x_num, num_results[t_val][var], 'bo-', markersize=0, linewidth=2, label='FVM (Lax-Friedrichs)')
        plt.plot(x_pinn, pinn_results[t_val][var], 'r-', linewidth=2, label='PINN')

        # If this is t=0, also plot the exact initial condition
        if t_val == 0.0:
            plt.plot(x_exact, exact_ic[var], 'k--', linewidth=2, label='Exact IC')

        plt.xlabel("x", fontsize=12)
        plt.ylabel(title, fontsize=12)
        plt.title(f"{title} at t = {t_val}", fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.legend(fontsize=10)

        plt.tight_layout()
        plt.savefig(f"{var}_comparison_at_t_{t_val}.png", dpi=300, bbox_inches="tight")
        plt.close()

# 2. Grid layout with all variables and time steps (more compact view)
fig, axes = plt.subplots(len(variables), len(time_points), figsize=(20, 12))
fig.tight_layout(pad=3.0)

for var_idx, (var, title) in enumerate(zip(variables, titles)):
    for t_idx, t_val in enumerate(time_points):
        ax = axes[var_idx, t_idx]

        # Plot both solutions
        ax.plot(x_num, num_results[t_val][var], 'bo-', markersize=0, linewidth=2, label='FVM')
        ax.plot(x_pinn, pinn_results[t_val][var], 'r-', linewidth=2, label='PINN')

        # If this is t=0, also plot the exact initial condition
        if t_val == 0.0:
            ax.plot(x_exact, exact_ic[var], 'k--', linewidth=2, label='Exact IC')

        # Add title only to top row
        if var_idx == 0:
            ax.set_title(f"t = {t_val}", fontsize=12)

        # Add y-label only to leftmost column
        if t_idx == 0:
            ax.set_ylabel(title, fontsize=12)

        # Add x-label only to bottom row
        if var_idx == len(variables) - 1:
            ax.set_xlabel("x", fontsize=10)

        # Add legend only to the first plot
        if var_idx == 0 and t_idx == 0:
            ax.legend(fontsize=8)

        ax.grid(True, alpha=0.3)

plt.suptitle("Comparison of FVM (Lax-Friedrichs) and PINN Solutions", fontsize=16)
plt.savefig("combined_comparison_grid.png", dpi=300, bbox_inches="tight")
plt.close()

# 3. Evolution plot for each variable (all time steps in one plot)
cmap = get_cmap('viridis')
time_colors = [cmap(i/len(time_points)) for i in range(len(time_points))]

for var_idx, (var, title) in enumerate(zip(variables, titles)):
    plt.figure(figsize=(12, 8))

    # Plot both solutions for all time points
    for i, t_val in enumerate(time_points):
        plt.plot(x_num, num_results[t_val][var], '-', color=time_colors[i], linewidth
```

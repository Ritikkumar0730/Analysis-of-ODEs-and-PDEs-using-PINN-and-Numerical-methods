```python
from astroquery.simbad import Simbad
import pandas as pd

# Input and output files
input_file = "/content/10_complete_sources.txt"
output_file = "output.csv"

# Read sources from text file (one per line)
with open(input_file, "r") as f:
    sources = [line.strip() for line in f if line.strip()]

# Prepare results list
results = []

for obj in sources:
    try:
        query = Simbad.query_object(obj)
        if query is not None:
            ra = query['RA'][0]   # RA in HMS
            dec = query['DEC'][0] # DEC in DMS
            results.append({"Name": obj, "RA": ra, "DEC": dec})
        else:
            results.append({"Name": obj, "RA": "Not found", "DEC": "Not found"})
    except Exception as e:
        results.append({"Name": obj, "RA": f"Error: {e}", "DEC": f"Error: {e}"})

# Convert to DataFrame and save to CSV
df = pd.DataFrame(results)
df.to_csv(output_file, index=False)

print(f"Saved results to {output_file}")
```
